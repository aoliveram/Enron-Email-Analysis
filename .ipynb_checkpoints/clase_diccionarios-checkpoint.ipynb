```python
import pandas as pd
import re
import string
import stanza

stanza.download('es')
nlp = stanza.Pipeline(lang='es', processors='tokenize,mwt,lemma') # Inicializar el pipeline de Stanza

def lemmatizer(text):
    doc = nlp(text)
    lemmatized_text = ' '.join([word.lemma for sentence in doc.sentences for word in sentence.words])
    return lemmatized_text

cabildos['fundamento'] = cabildos['fundamento'].str.lower() # Convertimos 'fundamento' a minúsculas
cabildos['fundamento'] = cabildos['fundamento'].apply(lambda x: re.sub("[0-9]", '', x)) # Eliminamos dígitos
cabildos['fundamento'] = cabildos['fundamento'].str.translate(str.maketrans('', '', string.punctuation)) # Eliminamos puntuación
cabildos['fundamento'] = cabildos['fundamento'].str.strip() # Eliminamos espacios en blanco al inicio y al final del texto
cabildos['fundamento'] = cabildos['fundamento'].str.replace('\n', ' ') # Reemplazamos saltos de línea por espacios


cabildos = cabildos[[len(x.split()) > 2 for x in cabildos['fundamento']]] # Filtramos donde 'fundamento' tiene más de 2 palabras
cabildos['fundamento'] = cabildos['fundamento'].apply(lemmatizer) # Aplicamos la lematización

# Seleccionamos aleatoriamente 1000 filas
cabildos = cabildos.sample(1000)

# Reiniciamos el índice del DataFrame y renombramos la columna de índice a 'id'
cabildos = cabildos.reset_index().rename(columns={'index': 'id'})
```

```python
df_liwc=pd.DataFrame()


for ix in range(cabildos.shape[0]):
  texto = cabildos.fundamento[ix].lower()
  tokens = word_tokenize(texto)
  tcounts1 = Counter(category for token in tokens for category in parse(token))
  df_counts = pd.DataFrame.from_dict([tcounts1]).reset_index()
  df_counts['index']=ix
  df_liwc = pd.concat([df_liwc,df_counts], axis=0, ignore_index=True)  

df_liwc = df_liwc.fillna(0)
```

```python
# cargando el modelo pre-entrenado y el diccionario
model_path='sbw_vectors.bin'
model, num_features, model_word_set=ddr.load_model(model_path)
dic_terms = ddr.terms_from_liwc(input_path = 'mfd_es_test.dic')

# definiendo directorios:
dictionary_vector_path='agg_dic_vectors.tsv'
document_vector_path='agg_doc_vecs.tsv'
document_loadings_out_path='document_dictionary_loadings.tsv'

# en esta parte, el diccionario se cruza con el modelo, para tener la representación vectorial de cada diccionario
# OJO ACA NO LEE STEMS!!!!
agg_dic_vecs = ddr.dic_vecs(dic_terms = dic_terms, model = model, num_features = num_features, model_word_set =model_word_set )

# exportando la representación del diccionario a CSV
ddr.terms_to_csv(agg_dic_vecs, output_path = dictionary_vector_path, delimiter ="\t")
```

```python
textos_cabildos = cabildos[['fundamento','id']]

# esto es importate, limpiar de saltos de linea, porque si no quedan más documentos, y no calzan los id
textos_cabildos['fundamento']=textos_cabildos['fundamento'].str.lower().str.replace('\n\n', '').str.replace('\n', '').str.replace("'","").str.replace('"', '').str.replace('”', '').str.replace('“', '').str.replace('¿', '').str.replace('?', '')

# crea el archivo csv de documentos
textos_cabildos.to_csv('textos_cabildos_ddr.csv', mode='w', columns=['fundamento','id'],encoding='utf-8',sep="\t", index=False)

# cargando textos 
documents_path='textos_cabildos_ddr.csv'

ddr.doc_vecs_from_csv(input_path = documents_path, output_path = document_vector_path, model = model, num_features = num_features, quotechar = None, model_word_set = model_word_set, header = True, text_col='fundamento' , delimiter="\t", id_col='id')
```

```python
with open(document_vector_path) as input, open('auxiliar.tsv', 'w') as output:
    non_blank = (line for line in input if line.strip())
    output.writelines(non_blank)

document_vector_path='auxiliar.tsv' 
```

```python
asd = ddr.get_loadings(agg_doc_vecs_path=document_vector_path, agg_dic_vecs_path=dictionary_vector_path, out_path=document_loadings_out_path,num_features=num_features,delimiter ="\t")

```
